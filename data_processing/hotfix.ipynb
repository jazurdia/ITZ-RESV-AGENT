```python
import pandas as pd
from decimal import Decimal
```

```python
df = pd.read_excel("data/itzana_reservations.xlsx")
```

```python
df.head()
```

```python
columns_TIME = ["ARRIVAL_TIME" , "DEPARTURE_TIME"]
columns_DATE = ["ARRIVAL", "DEPARTURE"]
columns_real = ["ROOM_NO", "LAST_ROOM"]
```

```python
def transform_df(df):
    """
    Transforms the DataFrame according to the specified column groups:
    - columns_TIME: convert to datetime.time
    - columns_DATE: convert to 'YYYY-MM-DD' strings
    - columns_real: ensure float dtype
    - other columns: convert numeric values to Decimal
    """
    # Convert time columns to datetime.time
    for col in columns_TIME:
        df[col] = pd.to_datetime(df[col], errors='coerce').dt.time

    # Convert date columns to ISO format 'YYYY-MM-DD'
    for col in columns_DATE:
        df[col] = pd.to_datetime(df[col], errors='coerce').dt.strftime('%Y-%m-%d')

    # Ensure real number columns remain floats
    for col in columns_real:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)

    # Identify and transform all other columns
    other_cols = [c for c in df.columns if c not in set(columns_TIME + columns_DATE + columns_real)]
    for col in other_cols:
        df[col] = df[col].apply(
            lambda x: Decimal(str(x)) if isinstance(x, (int, float)) else x
        )

    return df
```

```python
new_df = transform_df(df)
```

```python
new_df
```

```python
# Elimina los primeros tres caracteres de cada calor de la columna COMPANY_NAME

new_df["COMPANY_NAME"] = new_df["COMPANY_NAME"].astype(str).str[3:]
```
